```python
def get_wall_info(posts, other_posts):
    for i in range(len(other_posts['items'])):
        posts.loc[i] = ''
        posts.loc[i]['index'] = i
        posts.loc[i]['from_id'] = other_posts['items'][i]['from_id']
        posts.loc[i]['post_id'] = other_posts['items'][i]['id']
        if (other_posts['items'][i].get('views') != None):
            posts.loc[i]['views'] = other_posts['items'][i]['views']['count']
        posts.loc[i]['text'] = other_posts['items'][i]['text']
        posts.loc[i]['comments'] = other_posts['items'][i]['comments']['count']
        posts.loc[i]['likes'] = other_posts['items'][i]['likes']['count']
        posts.loc[i]['reposts'] = other_posts['items'][i]['reposts']['count']
        if (other_posts['items'][i].get('copy_history') != None):
            posts.loc[i]['is_repost'] = 1
            posts.loc[i]['repost_text'] = other_posts['items'][i]['copy_history'][0]

        else:
            posts.loc[i]['is_repost'] = 0
    return posts
```
